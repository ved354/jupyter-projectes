```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.base import BaseEstimator,TransformerMixin
from itertools import combinations
```

```python
##retreving data from my directry by pd.read_csv()
data=pd.read_csv("/home/vedanth-sai-ram/Downloads/train.csv")
```

```python
#converting my data into pandas dataframes by pd.DataFrame()
data__=pd.DataFrame(data)
```

```python
#trimimg the usefull data
data_trimed=data__.loc[:,["Sex","Pclass","Age","SibSp","Parch","Fare","Survived","Embarked"]]
```

```python
# data_trimed.info()
names_of_col=["Pclass","Parch","SibSp","Embarked"]
combos=[list(c) for c in (combinations(names_of_col,2))]
# combos
```

```python
fixed_col=["Age","Sex","Fare"]
class adder_transformer(BaseEstimator,TransformerMixin):
    def __init__(self,combo_col=None):
        self.fixed_col=fixed_col
        self.combo_col=combo_col
    def fit(self,x,y=None):
        return self
    def transform(self, x):
        if self.combo_col is not None:
            print(self.combo_col)
            print(self.fixed_col)
            x_copy=pd.DataFrame(x).copy()
            total_col=list(self.fixed_col)+list(self.combo_col)
            return x_copy[total_col]
```

```python
age_filler=Pipeline(steps=[
    ("age_filler",SimpleImputer(strategy="mean")),
    ("age_con",MinMaxScaler())
])
cat_conver=Pipeline(steps=[
    ("sex_con",OneHotEncoder()),
])
converter=Pipeline(steps=[
    ("val_con",MinMaxScaler()),
])
```

```python
preprocessing=ColumnTransformer(transformers=[
    ("encoder",cat_conver,["Sex"]),
    ("filler",age_filler,["Age"]),
    ("scaler",converter,["Fare"]),
],remainder='passthrough')
```

```python
full_pipeline=Pipeline([
    ("adder",adder_transformer()),
    ("pip_1",preprocessing),
    ("final_pred",LogisticRegression())
])
```

```python
pragma={
    "adder__combo_col":combos,
}
```

```python
grid=GridSearchCV(full_pipeline,pragma,cv=8)
```

```python
train_x,test_x,train_y,test_y=train_test_split(data_trimed.loc[:,["Sex","Age","Fare","Pclass","SibSp","Parch","Embarked"]],data_trimed.loc[:,["Survived"]],test_size=0.1,random_state=42)
```

```python
grid.fit(train_x,train_y)
```

```python
print(grid.best_params_)   # e.g. {'adder__combo': ['Pclass', 'SibSp']}
print(grid.best_score_)
```

```python
pred=grid.predict(test_x)
pred
```

```python
corr=test_y.to_numpy().reshape(pred.shape)
```

```python
correct=corr-pred
```

```python
print((np.sum(correct==0)/len(correct))*100)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(test_y,pred)
```
