```python
import math
data=np.random.normal(0,10,50)
data_redi=[]
values=[]
vales_2=[]
for i in data:
    if i>0:
        data_redi.append(i)
        see=math.log(i)
        values.append(see)
for i in values:
    see=math.exp(i)
    vales_2.append(see)
    
```

```python
plt.scatter(data_redi,values)
plt.scatter(values,vales_2)
plt.show()
```

```python
plt.scatter(values,vales_2)
```

```python
e_curve=[]
```

```python
for i in data_redi:
    formula=1/(1+math.exp(-i))
    e_curve.append(formula)
```

```python
plt.scatter(data_redi,e_curve)
```

```python
from sklearn import datasets
irirs=datasets.load_iris()
#irirs.keys()
data=irirs["data"][:,3:]
output=(irirs["target"]==2)
out_num=[int(i) for i in output]
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(data,out_num)
a=np.linspace(0,3,100).reshape(-1,1)
predic=model.predict_proba(a)
predict_2=model.predict(a)
```
