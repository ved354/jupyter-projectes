```python
import sys

# Remove system path
sys.path = [p for p in sys.path if '/usr/lib/python3' not in p]

# Remove already-cached mpl_toolkits from memory
mods_to_remove = [key for key in sys.modules if 'mpl_toolkits' in key or 'matplotlib' in key]
for mod in mods_to_remove:
    del sys.modules[mod]

# Now import fresh
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
print("Works!", ax)
```
