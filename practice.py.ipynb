```python
import math
data=np.random.normal(0,10,50)
data_redi=[]
values=[]
vales_2=[]
for i in data:
    if i>0:
        data_redi.append(i)
        see=math.log(i)
        values.append(see)
for i in values:
    see=math.exp(i)
    vales_2.append(see)
    
```

```python
plt.scatter(data_redi,values)
plt.scatter(values,vales_2)
plt.show()
```

```python
from sklearn import datasets
irirs=datasets.load_iris()
#irirs.keys()
data=irirs["data"][:,3:]
output=(irirs["target"]==2)
out_num=[int(i) for i in output]
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(data,out_num)
a=np.linspace(0,3,100).reshape(-1,1)
predic=model.predict_proba(a)
predict_2=model.predict(a)
```

```python
from sklearn.datasets import make_moons
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
import matplotlib.pyplot as plt
```

```python
x,y=make_moons(n_samples=100,shuffle=True,noise=0.12,random_state=42)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
work_flow=Pipeline([
    ("pliy_data",PolynomialFeatures(degree=2)),
     ("scaler",StandardScaler()),
     ("model", LinearSVC(C=10,loss="hinge")),
     ])
```

```python
work_flow.fit(x_train,y_train)
```
