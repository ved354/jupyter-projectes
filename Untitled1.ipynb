```python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score,recall_score
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import roc_curve
from sklearn.multiclass import OneVsOneClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC
import numpy as np
import pandas as pd
```

```python

data=fetch_openml('mnist_784',version=1)
main_data=data["data"]
main_lables=data["target"]
x_train,x_test,y_train,y_test=train_test_split(main_data,main_lables,test_size=0.2,random_state=42)
```

```python
pro_split=StratifiedKFold(n_splits=3,random_state=42,shuffle=True)
main_lables_2=y_train=="5"
```

```python
pro_split=StratifiedKFold(n_splits=3,shuffle=True,random_state=42)
```

```python

svm_clasifier=SVC()
svm_clasifier.fit(x_train,y_train)
```

```python
cross_val_predict_2=cross_val_predict(svm_clasifier,x_train.values,y_train.values,cv=pro_split)
```

```python
cla_a=3
cla_b=5
real_aa=x_train[(cross_val_predict_2==cla_a)&(y_train==cla_a)]
dub_ab=x_train[(cross_val_predict_2==cla_b)&(y_train==cla_a)]
real_bb=x_train[(cross_val_predict_2==cla_b)&(y_train==cla_b)]
dub_ba=x_train[(cross_val_predict_2==cla_a)&(y_train==cla_b)]



def plot_digits(insetense,images_per_row):
    size=28
    image_per_row_colums=min(len(insetense),images_per_row)
    n_rows=(len(insetense)-1)//(images_per_row)+1 
    empty_slotes=images_per_row*n_rows-len(insetense)
    patched_instense=np.concatenate((insetense,np.zeros((empty_slotes,size*size))),axis=0)
    row_imges_gen=[]
    for row in range(n_rows):
        rimages=patched_instense[row*images_per_row:(row+1)*images_per_row]
        row_imges_gen.append(np.concatenate((rimages.reshape(-1,28)),axis=1))
    print(row_imges_gen)
    final_image=np.vstack(row_imges_gen)
    plt.imshow(final_image,cmap="binary")
    plt.axis("off")

plt.figure(figsize=(8,8))
plt.subplot(221);plot_digits(real_aa[:25],images_per_row=5)
plt.subplot(222);plot_digits(dub_ab[:25],images_per_row=5)
plt.subplot(223);plot_digits(real_bb[:25],images_per_row=5)
plt.subplot(224);plot_digits(dub_ba[:25],images_per_row=5)
plt.show()
```
